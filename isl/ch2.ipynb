```python
# different types of random matrices
matrix_rnd_normal = np.random.normal(0, 1, (1, 1000))
matrix_rnd_uniform = np.random.rand(1, 1000)
```

```python
print(matrix_rnd_normal.mean(axis=1).item(), matrix_rnd_normal.std(axis=1).item()) # should be (0, 1)
print(matrix_rnd_uniform.mean(axis=1).item(), matrix_rnd_uniform.var(axis=1).item()) # should be 0.5, 1/12
```

```python
# (x: uniform, y: normal)
plt.figure(figsize=(10, 6))

normal_arr = matrix_rnd_normal.T.squeeze()
uniform_arr = matrix_rnd_uniform.T.squeeze()

plt.scatter(uniform_arr, normal_arr)
```

```python
# histograms
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.hist(normal_arr, bins=20)
plt.subplot(122)
plt.hist(uniform_arr, bins=20)
```
